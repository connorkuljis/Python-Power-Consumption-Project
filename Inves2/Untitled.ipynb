```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv

appliance = []
wattage = []
usage = []
with open('valerie.csv', newline='') as csvfile:
    modelreader = csv.reader(csvfile, delimiter=':')
    for row in modelreader:
        appliancenames = row[0]
        appliance.append(appliancenames)
        
        appliancewattage = float(row[1])
        wattage.append(appliancewattage)
                       
        data = row[2]
        # list comprehension splitting the long string of numbers into floats and multiplying by wattage
        nicedata = [float(i) * appliancewattage for i in data.split(',')] 
        usage.append(nicedata)

times=range(1,25)
plt.stackplot(times, usage, labels=appliance)
plt.show()
```
